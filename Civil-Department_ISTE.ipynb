```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
values=[48,41]
y=values
color=['red','lime']
labels=['Students Affected in civil','Normal students']
ex=[0,0.1]
def make_autopct(values):
    def my_autopct(pct):
        total = sum(values)
        val=int((pct*total/100.0)+0.5)
        return '{p:.2f}%  ({v:d})'.format(p=pct,v=val)
    return my_autopct
plt.pie(values,labels=labels,colors=color,explode=ex,autopct=make_autopct(y))
```
