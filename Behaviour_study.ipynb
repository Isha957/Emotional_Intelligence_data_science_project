```python
import numpy as np
import matplotlib.pyplot as plt


empathy = pd.DataFrame([13, 38, 40, 41, 48, 49,41,35,43,42,28,48,35,45,37,49,43,36,40,46,41,37,46,43,36,33,43,41,27,41,42,46,38,41,26,37,42,39,38,39,34,40,43,41,30,41,39,35,50,36,35], columns=['Empathy'])

x = np.arange(len(empathy.index))
y = empathy['Empathy']
fig = plt.figure()
fig.set_figheight(4)
fig.set_figwidth(15)
mask1 = y <= 35
mask2 = y > 35

plt.bar(x[mask1], y[mask1], color = 'green')
plt.bar(x[mask2], y[mask2], color = 'yellow')
plt.xlabel("Questions")
plt.ylabel("Score")

plt.gca().legend(('score<35','score>35'))

plt.show()
  
    
```

##### In Above graph Green represent students having Empathy  below 35 and yellow vice-versa
