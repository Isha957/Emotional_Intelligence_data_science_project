```python

# This code below is to find all students who are having social weakness there roll no are listed at right in the legend

import matplotlib.pyplot as plt
%matplotlib inline
sk = pd.DataFrame(sk['social'], columns=['social'])
x = np.arange(len(sk.index))
y = sk['social']
colors = ['#f22c2c','#f2552c','#f2c02c','#8ff22c','#2cf22c','#2cf25d','#2cf2c0','#2c8ff2','#c02cf2','#f22c5d','#f22c2c','#7b4500']
values = [3, 12, 5, 8] 
def make_autopct(values):
    def my_autopct(pct):
        total = sum(values)
        val=int((pct*total/100.0)+0.5)
        return '{p:.2f}%  ({v:d})'.format(p=pct,v=val)
    return my_autopct
mask1 = y <= 35
mask2 = y > 35
plt.pie( y[mask1],autopct=make_autopct(y[mask1]),colors=colors, pctdistance=2.0,labeldistance=0.5, startangle=90,counterclock=False)
plt.axis('equal')
plt.legend(x[mask1],loc="upper left")
plt.title("Here legend means student rollno's")
plt.show()

```
